# 1670. Design Front Middle Back Queue (Medium)
Design a queue that supports push and pop operations in the front, middle, and back.<br/>
Implement the FrontMiddleBack class:
- FrontMiddleBack() Initializes the queue.
- void pushFront(int val) Adds val to the front of the queue.
- void pushMiddle(int val) Adds val to the middle of the queue.
- void pushBack(int val) Adds val to the back of the queue.
- int popFront() Removes the front element of the queue and returns it. If the queue is empty, return -1.
- int popMiddle() Removes the middle element of the queue and returns it. If the queue is empty, return -1.
- int popBack() Removes the back element of the queue and returns it. If the queue is empty, return -1.
Notice that when there are two middle position choices, the operation is performed on the frontmost middle position choice.<br/> 
#
For example:
- Pushing 6 into the middle of [1, 2, 3, 4, 5] results in [1, 2, 6, 3, 4, 5].
- Popping the middle from [1, 2, 3, 4, 5, 6] returns 3 and results in [1, 2, 4, 5, 6].

### Examples
Input:<br/>
["FrontMiddleBackQueue", "pushFront", "pushBack", "pushMiddle", "pushMiddle", "popFront", "popMiddle", "popMiddle", "popBack", "popFront"]<br/>
[[], [1], [2], [3], [4], [], [], [], [], []]<br/>
Output:<br/>
[null, null, null, null, null, 1, 3, 4, 2, -1]<br/>

Explanation:
- FrontMiddleBackQueue q = new FrontMiddleBackQueue();
- q.pushFront(1);   // [1]
- q.pushBack(2);    // [1, 2]
- q.pushMiddle(3);  // [1, 3, 2]
- q.pushMiddle(4);  // [1, 4, 3, 2]
- q.popFront();     // return 1 -> [4, 3, 2]
- q.popMiddle();    // return 3 -> [4, 2]
- q.popMiddle();    // return 4 -> [2]
- q.popBack();      // return 2 -> []
- q.popFront();     // return -1 -> [] (The queue is empty)

```python
class FrontMiddleBackQueue(object):
    def __init__(self):
        self.q = []

    def pushFront(self, val):
        """
        :type val: int
        :rtype: None
        """
        self.q.insert(0, val)
    
    def getQ(self):
        return self.q
    
    def pushMiddle(self, val):
        """
        :type val: int
        :rtype: None
        """
        self.q.insert(len(self.q) // 2, val)

    def pushBack(self, val):
        """
        :type val: int
        :rtype: None
        """
        self.q.append(val)

    def popFront(self):
        """
        :rtype: int
        """
        if self.isEmpty():
            return -1
        else:
           return self.q.pop(0)

    def popMiddle(self):
        """
        :rtype: int
        """
        if self.isEmpty():
            return -1
        else:
            if len(self.q) % 2 == 0:
                return self.q.pop((len(self.q) // 2 ) - 1)
            else:
                return self.q.pop(len(self.q) // 2)
        
    def popBack(self):
        """
        :rtype: int
        """
        if self.isEmpty():
            return -1
        else:
            return self.q.pop()
    
    def isEmpty(self):
        return len(self.q) == 0

obj = FrontMiddleBackQueue()

obj.pushFront(1)
assert obj.getQ() == [1]

obj.pushBack(2)
assert obj.getQ() == [1,2]

obj.pushMiddle(3)
assert obj.getQ() == [1,3,2]

obj.pushMiddle(4)
assert obj.getQ() == [1,4,3,2]

assert obj.popFront() == 1
assert obj.popMiddle() == 3
assert obj.popMiddle() == 4
assert obj.popBack() == 2
assert obj.popFront() == -1
```
